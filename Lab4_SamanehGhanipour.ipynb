# Question 6

Data preprations like text mining, deleting unnecessarie junctions, tokenization, lexicon normaliztion,making lower case of words,...

```python
from nltk.corpus import stopwords
import nltk as nlp
import nltk
import re
```

Here I am doing the text mining techniques like lexicon normalization,tokenizing, deleting stopwords,... 

```python
PreparedData=PreparedData.drop(0)
```

```python
description_list=[]
for description in PreparedData.dialogue:
    description=re.sub("[^a-zA-Z]", " ", description)
    description=description.lower()
    description=nltk.word_tokenize(description)
    description=[word for word in description if not word in set(stopwords.words("english"))]
    lemma=nlp.WordNetLemmatizer()
    description=[lemma.lemmatize(word) for word in description]
    description=" ".join(description)
    description_list.append(description)
```

Sentiment Analysis library which takes in a string and returns a dictionary of scores in each of the categories

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
sentiAna = SentimentIntensityAnalyzer()
```

```python
sentiAna
```

```python
New
```

I used Lambda to execute the expression

```python
PreparedData['sentiment_scores'] = PreparedData['dialogue'].apply(lambda dialogue: 
                                         sentiAna.polarity_scores(dialogue))
```

```python
PreparedData['compound_score']  = PreparedData['sentiment_scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
PreparedData['compound_sentiment'] = PreparedData['compound_score'].apply(lambda c: 'Possitive' if c >=0 else 'Negative')
```
